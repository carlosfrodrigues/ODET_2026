```python
df = df[df['tasmin'] != -999]
df["tasmax"] = df["tasmin"]*2
print(df)
```

```python
df.plot(x="year", y=["tasmin", "tasmax"], marker="x", xlabel="ano", ylabel="anomalia", grid=True, color=["black", "red"])
```
